```python
#自訂公式函數
#開啟並整理檔案
def o_f(file_name):
  a = str(file_name)+'.csv'
  csv = open(a, 'r',  encoding='utf-8')
  price_name = csv.readlines()
  csv.close()
  number = "".join(price_name)
  listcsv = number.split(',') 
  p1 = listcsv.pop(0)
  p2 = listcsv.pop(-1)
  num = []
  for number in listcsv:
    num.append(float(number))
  return num

#用來確定他在不在字典裡面
def Search(keyword, dictt):
  if keyword in dictt:
    return GetValue(keyword, dictt)
  else:
    print("此項指標不存在")
    nkey=input('請使用者重新輸入想要查詢的指標')
    return Search(nkey, dictt) #在第一個函數裡面叫第二個函數

#用來查詢後排序
def GetValue(key, dictt):
  names=['台積電','中華電','台塑','統一','萬海']
  readout=dictt.get(key)
  newdict={}
  n=0
  while n<len(readout):
    newdict[readout[n]]=names[n]
    n+=1
  nreadout=sorted(readout)
  nreadout.reverse()
  olist=[]
  n=0
  for key in nreadout:
    olist=olist+[newdict.get(key)]
  final=[]
  while n<len(nreadout):
    final=final+[olist[n]+':'+str(nreadout[n])]
    n+=1
  print(final)
  return(olist)

#差值
def de(list_name):
  de_list = []
  for number in list_name:
    de_list.append(number - sum(list_name)/len(list_name))
  return de_list

#標準差
def sd(list_name):
  sd_list = []
  for deviation in de(list_name):
    sd_list.append(deviation**2)
  return round((sum(sd_list)/(len(sd_list)-1))**(0.5),4)

#共變異數
def cov(list_A,list_B):
  cov_list = []
  for nA,nB in zip(de(list_A),de(list_B)):
    cov_list.append(nA*nB)
  return round(sum(cov_list)/len(cov_list),4)

#月報酬
def ret(list_name):
  ret = []
  for i in range(11):
    ret.append((list_name[i+1] - list_name[i]) / list_name[i])
  return ret

#Beta
def beta(return_name):
  B = round(cov(ret(return_name),ret(nummarket))/(sd(ret(nummarket))**2),4)
  return B

#股價預期報酬率
def er(return_name):
  rfr = 0.0057  #台灣十年期公債利率 #risk_free_return
  mr = round(sum(ret(nummarket))/len(ret(nummarket)),4)  #market return
  expected_return = rfr+beta(return_name)*(mr-rfr)
  return expected_return

#投資組合風險
def port(list_A,list_B):
  port_list = []
  sd1 = sd(list_A)
  sd2 = sd(list_B)
  covariance = cov(list_A,list_B)
  for i in range(0,100,1):
    j = 100 - i
    p = ((i/100)**2)*(sd1**2)+((j/100)**2)*(sd2**2)+2*(i/100)*(j/100)*covariance
    port_list.append((p**0.5))
  return port_list

#投資組合報酬率
def p_ret(list_A, list_B):
  ret_list = []
  er1 = er(list_A)
  er2 = er(list_B)
  for i in range(0,101,1):
    j = 100 - i
    r = (i/100)*er1 + (j/100)*er2
    ret_list.append(r)
  return ret_list

#尋找最佳組合點
def b_point(list_A, list_B):
  bet = min(port(list_A, list_B))
  for i in port(list_A, list_B):
    if i == bet:
      a = port(list_A, list_B).index(i)
  bett = p_ret(list_A, list_B)[a]
  weight = a
  print('最佳投資組合比例為：第一支股票',weight,'%','、第二支股票',100-weight,'%；投資報酬率為：',bett,'；風險為',bet)
```
